```python
x = 10  # Global variable

def update_x():
    global x  # Declare we suing global variable x
    x += 5    
   
update_x()
print(x)  # Output: 15
```

```python
x = 10  # Global variable

def update_x():
    globals()['x'] += 5  # Access & modify  global variable 

update_x()
print(x)  # Output: 15
```
